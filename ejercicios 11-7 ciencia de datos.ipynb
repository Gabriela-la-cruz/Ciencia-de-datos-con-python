PRACTICA 21: partiendo de la lista vip= [18,35,78,44,63,24,88,125,7,53] 
lo convertiremos a una serie de pandas y mediante una función restaremos 3 a todos los elementos cuyos valores sean menores de 50
queremos ver en un texto el total de la suma, antes y despues de aplicar la función 
y en otro texto la diferencia entre los dos totales 

```python
import pandas as pd 
vip = pd.Series([18,35,78,44,63,24,88,125,7,53]) 
antes = pd.Series(data=vip)

def resta3(x):
    if x < 50:
         return x-3
    else:
         return x
despues = antes.apply(resta3)
vip.apply(resta3) # asi esta bien pero no sirve para hacer la resta de los resultados 
print("la suma del antes es", antes.sum(),"la suma del despues es", despues.sum(), "la diferencia es", antes.sum()-despues.sum())
```

```python
# como ordenar uan serie 
# etl es la fase en la que se transforman los datos, sin transformar los datos no podemos interpretarlos correctamente 
#ignore_index=True esto lo que hace es ordenar el indice tambien por orden ascendente (reindexado)
antes.sort_values(ascending = True, ignore_index=True)
# modificaciones principales en una serie 
#para que haya una modificacion tiene que haber un = 
antes = antes.sort_values(ascending = True, ignore_index=True)
```

```python
#como filtrar que solo me muestre los mayores de 40
antes[antes>40].sort_values(ascending=True, ignore_index=True)
```

```python
#PRACTICA 22:
#CREAREMOS UNA SERIE NUMERICA DE PANDAS CON ESTOS VALORES [34,124,81,9,27,81,34,63,75,44,18,67,22,63,25,63] ESTA SERIE SE LLAMARA NUMEROS VIP 
#CLONAREMOS LA SERIE PARA TENER SOLAMENTE LOS VALORES MAYORES DE 40 QUE ESTE EN ORDEN DESC Y REINDEXADO, ESTA NUEVA SERIE SE LLAMARA NUMEROS VIP 40
# A ESTA NUEVA SERIE LE APLICAREMOS UNA FUNCION QUE RESTE 2 SI EL VALOR ES MAYOR DE 70 O SI ES MENOR O IGUAL A 70 SE SUMA 2 
# VER EN UN TEXTO LA VARIANZA LA DESVIACION Y LA SUMA DE NUMEROS VIP 40

import pandas as pd 
numerosvip = pd.Series([34,124,81,9,27,81,34,63,75,44,18,67,22,63,25,63])

vip40 = numerosvip [numerosvip > 40].sort_values(ascending = False, ignore_index=True)

def resta2(x): #se puede llamar de cualquier manera no solo resta2 ej: lafuncion(x) o otro nombre
    if x > 70:
        return x-2
    elif x <= 70: 
        return x + 2
vip40 = vip40.apply(resta2)

print("la suma acumulada es",round(vip40.cumsum(),2))
print("la desviacion tipica es", round(vip40.std(),2))
print("la varianza es", round(vip40.var(),2))

```
